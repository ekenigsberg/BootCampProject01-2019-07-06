```python
import pandas as pd
import numpy as np

```

```python

#Reading in CSV files pulled from the National Water Quality Monitoring Portal
temp12_file = "2012_temp_data.csv"
temp82_file = "1982_temp_data.csv"
temp12_pd = pd.read_csv(temp12_file)
temp82_pd = pd.read_csv(temp82_file)
temp12_pd.head()
```

```python
#Cleaning to the two data pulls to reduce dataframes to pertinent data only. 
#Accomplished by creating cleaned dataframes with specific columns
temp_12 = temp12_pd[["MonitoringLocationIdentifier","ActivityStartDate","CharacteristicName","ResultMeasureValue","ResultMeasure/MeasureUnitCode" ]]
temp_82 = temp82_pd[["MonitoringLocationIdentifier","ActivityStartDate","CharacteristicName","ResultMeasureValue","ResultMeasure/MeasureUnitCode" ]]
```

```python
#Displaying cleaned Dataframe

temp_82.head()
```

```python
#Dropping Null values from each dataframe

cleaned_temp12 = temp_12.dropna(how="any")
cleaned_temp82 = temp_82.dropna(how="any")
```

```python
#Creating new dataframes to add in "numeric" result values
num_temp12 = cleaned_temp12
num_temp82 = cleaned_temp82
num_temp82.head()
```

```python
#Cleaned data by making all results a numeric value, and extract only results that were recored in Celsius
#Over 95% of provided data was in Celsius
num_temp12["Result"] = pd.to_numeric(cleaned_temp12["ResultMeasureValue"],'coerce')
num_temp82["Result"] = pd.to_numeric(cleaned_temp82["ResultMeasureValue"],'coerce')
cels_results = num_temp12.loc[num_temp12["ResultMeasure/MeasureUnitCode"] == "deg C"]
cels_results2 = num_temp82.loc[num_temp82["ResultMeasure/MeasureUnitCode"] == "deg C"]
```

```python
#Grouping each Dataframe by the Monitoring Location ID, allowing us to determine a mean for the temperature of each location
        #within each of the dataframes
#Once the Dataframes are grouped, the mean for each group is calculated and placed dinside a new dataframe

grouped_12 = cels_results.groupby(["MonitoringLocationIdentifier"])
grouped_82 = cels_results2.groupby(["MonitoringLocationIdentifier"])
mean_by_ID = grouped_12["Result"].mean()
mean_by_ID2 = grouped_82["Result"].mean()
mean_pd = pd.DataFrame(mean_by_ID)
mean_pd2 = pd.DataFrame(mean_by_ID2)
mean_pd2.head()
```

```python
#Drop any additional null values from the final results and output to a CSV.

cleaned_mean = mean_pd.dropna(how="any")
cleaned_mean2 = mean_pd2.dropna(how="any")

cleaned_mean.to_csv("Mean_by_ID_2012.csv",index=True, header=True)
cleaned_mean2.to_csv("Mean_by_ID_1982.csv",index=True, header=True)
```
