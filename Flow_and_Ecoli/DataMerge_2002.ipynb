```python
import pandas as pd
import numpy as np
import datetime
```

```python
cc_csv = '../State_County_Code_Data.csv'
mli_csv = '../../MLI2.csv'
data_2002 = '../../narrowresult 5.csv'

county_cd = pd.read_csv(cc_csv)
mli = pd.read_csv(mli_csv)
data_2002 = pd.read_csv(data_2002)
mli.info()
#data_2002.info()
```

```python
clean_data_2002 = data_2002.dropna(axis=1, how='all')



new_data_2002 = clean_data_2002[['OrganizationIdentifier', 'OrganizationFormalName',
       'ActivityIdentifier', 'ActivityStartDate', 'ActivityStartTime/Time',
       'ActivityStartTime/TimeZoneCode', 'MonitoringLocationIdentifier',
       'ResultIdentifier','CharacteristicName', 'ResultMeasureValue',
       'ResultMeasure/MeasureUnitCode', 'ResultStatusIdentifier', 'ResultValueTypeName','USGSPCode','ProviderName']]


data_2002_df = new_data_2002.dropna(axis=0, how='any')
data_2002_df
```

```python
#mli = mli[mli['OrganizationIdentifier'].astype(str)]
mli = mli[mli['OrganizationIdentifier'] != 'ARS']

mli_df = mli[['OrganizationIdentifier', 'OrganizationFormalName',
       'MonitoringLocationIdentifier', 'LatitudeMeasure', 'LongitudeMeasure',
       'StateCode', 'CountyCode']]
mli_df
```

```python
combined_2002_df = pd.merge(data_2002_df, mli_df, on='MonitoringLocationIdentifier', how='left')

data_2002_df = combined_2002_df[['OrganizationIdentifier_x', 'OrganizationFormalName_x',
    'ActivityStartDate', 'ActivityStartTime/Time',
       'ActivityStartTime/TimeZoneCode', 'MonitoringLocationIdentifier',
       'CharacteristicName', 'ResultMeasureValue',
       'ResultMeasure/MeasureUnitCode', 'ResultStatusIdentifier',
       'ResultValueTypeName', 'LatitudeMeasure', 'LongitudeMeasure', 'StateCode', 'CountyCode']]

data_2002_df = data_2002_df.rename(columns={'OrganizationIdentifier_x': 'Org Identifier', 'OrganizationFormalName_x': 'Org Name',
    'ActivityStartDate': 'Start Date', 'ActivityStartTime/Time':'Start Time',
       'ActivityStartTime/TimeZoneCode': 'Time Zone',
       'CharacteristicName': 'Characteristic Name', 'ResultMeasureValue': 'Result Value',
       'ResultMeasure/MeasureUnitCode': 'Unit Code', 'ResultStatusIdentifier': 'Result Status',
       'ResultValueTypeName': 'Result Value Type', 'LatitudeMeasure': 'Latitude', 'LongitudeMeasure':'Longitude'}
                                )
data_2002_df

data_2002_df = data_2002_df.astype({"StateCode":int, "CountyCode":int})


state_df = data_2002_df['State'] = data_2002_df['StateCode'].apply(lambda x: str(x).zfill(2))
state_df

county_df = data_2002_df['County'] = data_2002_df['CountyCode'].apply(lambda x: str(x).zfill(3))
county_df



FIPS_df = data_2002_df['FIPS'] = data_2002_df.State + data_2002_df.County
FIPS_df

data_2002_df.columns

data_2002_df = data_2002_df[['FIPS', 'Org Identifier', 'Org Name', 'Start Date', 'Start Time', 'Time Zone',
       'MonitoringLocationIdentifier', 'Characteristic Name', 'Result Value',
       'Unit Code', 'Result Status', 'Result Value Type', 'Latitude',
       'Longitude', 'StateCode', 'CountyCode']]

data_2002_df
```

```python
data_2002_df["Start Date"] = pd.to_datetime(data_2002_df["Start Date"])

data_2002_df
```

```python
data_2002_df.to_csv("2002_WQ_Data.csv", index=False)
```

```python
ph_df = data_2002_df.loc[data_2002_df['Characteristic Name'] == 'pH']
temp_df = data_2002_df.loc[data_2002_df['Characteristic Name'] == 'Temperature, water']
turb_df = data_2002_df.loc[data_2002_df['Characteristic Name'] == 'Turbidity']
fecal_df = data_2002_df.loc[data_2002_df['Characteristic Name'] == 'Fecal Coliform']
fecal_df
```

```python
mergedata_df = pd.merge(ph_df, temp_df, on='MonitoringLocationIdentifier', how='left')

mergedata2_df = pd.merge(turb_df, fecal_df, on='MonitoringLocationIdentifier', how='left')

merge_df = mergedata_df.columns

merge_df = mergedata_df[['FIPS_x', 'Org Identifier_x',
        'MonitoringLocationIdentifier',
       'Result Value_x','Result Value_y',
         'StateCode_x', 'CountyCode_x']]

merge_df = merge_df.rename(columns={'FIPS_x': 'FIPS', 'Org Identifier_x': 'Org',
       'Result Value_x': 'pH',
       'Result Value_y': 'Temperature',
         'StateCode_x': 'StateCode', 'CountyCode_x':'CountyCode'}
                          )

merge2_df = mergedata2_df[['FIPS_x', 'Org Identifier_x',
        'MonitoringLocationIdentifier',
       'Result Value_x','Result Value_y',
         'StateCode_x', 'CountyCode_x']]

merge2_df = merge2_df.rename(columns={'FIPS_x': 'FIPS', 'Org Identifier_x': 'Org',
       'Result Value_x': 'Turbidity',
       'Result Value_y': 'Fecal Coliform',
         'StateCode_x': 'StateCode', 'CountyCode_x':'CountyCode'}
                          )
merge_df
```

```python
merge_df = merge_df.astype({"pH":float, "Temperature":float})
merge2_df = merge2_df.astype({"Turbidity":float, "Fecal Coliform":float})

grp_merge_df = merge_df.groupby(['FIPS']).mean()

grp_merge2_df = merge2_df.groupby(['FIPS']).mean()

#grp_merge_df = grp_merge_df.mean(['pH'],['Temperature'])
grp_merge2_df

grp_merge_df.columns

grp_merge_df = grp_merge_df[['pH', 'Temperature']]
grp_merge2_df = grp_merge2_df[['Turbidity', 'Fecal Coliform']]

grp_merge_df
```

```python
comb_merge = pd.merge(grp_merge_df, grp_merge2_df, on='FIPS', how='left')
comb_merge
```

```python
comb_merge.to_csv("2002_MeanValue_Data.csv", index=True)
```
