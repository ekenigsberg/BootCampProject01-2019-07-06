```python
import pandas as pd
import numpy as np
import datetime
```

```python
cc_csv = '../State_County_Code_Data.csv'
mli_csv = '../../MLI2.csv'
flow_csv = 'Clean_Flow_Data.csv'
fecal_csv = 'Clean_Fecal_Coliform_Data.csv'

county_cd = pd.read_csv(cc_csv)
mli = pd.read_csv(mli_csv)
flow_df = pd.read_csv(flow_csv)
fecal_df = pd.read_csv(fecal_csv)
mli.info()
```

```python
#mli = mli[mli['OrganizationIdentifier'].astype(str)]
mli = mli[mli['OrganizationIdentifier'] != 'ARS']

mli_df = mli[['OrganizationIdentifier', 'OrganizationFormalName',
       'MonitoringLocationIdentifier', 'LatitudeMeasure', 'LongitudeMeasure',
       'StateCode', 'CountyCode']]
mli_df
```

```python
combined_flow_df = pd.merge(flow_df, mli_df, on='MonitoringLocationIdentifier', how='left')

new_flow_df = combined_flow_df[['OrganizationIdentifier_x', 'OrganizationFormalName_x',
    'ActivityStartDate', 'ActivityStartTime/Time',
       'ActivityStartTime/TimeZoneCode', 'MonitoringLocationIdentifier',
       'CharacteristicName', 'ResultMeasureValue',
       'ResultMeasure/MeasureUnitCode', 'ResultStatusIdentifier',
       'ResultValueTypeName', 'LatitudeMeasure', 'LongitudeMeasure', 'StateCode', 'CountyCode']]

new_flow_df = new_flow_df.rename(columns={'OrganizationIdentifier_x': 'Org Identifier', 'OrganizationFormalName_x': 'Org Name',
    'ActivityStartDate': 'Start Date', 'ActivityStartTime/Time':'Start Time',
       'ActivityStartTime/TimeZoneCode': 'Time Zone',
       'CharacteristicName': 'Characteristic Name', 'ResultMeasureValue': 'Result Value',
       'ResultMeasure/MeasureUnitCode': 'Unit Code', 'ResultStatusIdentifier': 'Result Status',
       'ResultValueTypeName': 'Result Value Type', 'LatitudeMeasure': 'Latitude', 'LongitudeMeasure':'Longitude'}
                                )
new_flow_df

new_flow_df = new_flow_df.astype({"StateCode":int, "CountyCode":int})


state_flow = new_flow_df['State'] = new_flow_df['StateCode'].apply(lambda x: str(x).zfill(2))
state_flow

county_flow = new_flow_df['County'] = new_flow_df['CountyCode'].apply(lambda x: str(x).zfill(3))
county_flow



FIPS_flow = new_flow_df['FIPS'] = new_flow_df.State + new_flow_df.County
FIPS_flow

new_flow_df.columns

new_flow_df = new_flow_df[['FIPS', 'Org Identifier', 'Org Name', 'Start Date', 'Start Time', 'Time Zone',
       'MonitoringLocationIdentifier', 'Characteristic Name', 'Result Value',
       'Unit Code', 'Result Status', 'Result Value Type', 'Latitude',
       'Longitude', 'StateCode', 'CountyCode']]

new_flow_df
```

```python
combined_fecal_df = pd.merge(fecal_df, mli_df, on='MonitoringLocationIdentifier', how='left')

new_fecal_df = combined_fecal_df[['OrganizationIdentifier_x', 'OrganizationFormalName_x',
    'ActivityStartDate', 'ActivityStartTime/Time',
       'ActivityStartTime/TimeZoneCode', 'MonitoringLocationIdentifier',
       'CharacteristicName', 'ResultMeasureValue',
       'ResultMeasure/MeasureUnitCode', 'ResultStatusIdentifier',
       'ResultValueTypeName', 'LatitudeMeasure', 'LongitudeMeasure', 'StateCode', 'CountyCode']]

new_fecal_df = new_fecal_df.rename(columns={'OrganizationIdentifier_x': 'Org Identifier', 'OrganizationFormalName_x': 'Org Name',
    'ActivityStartDate': 'Start Date', 'ActivityStartTime/Time':'Start Time',
       'ActivityStartTime/TimeZoneCode': 'Time Zone',
       'CharacteristicName': 'Characteristic Name', 'ResultMeasureValue': 'Result Value',
       'ResultMeasure/MeasureUnitCode': 'Unit Code', 'ResultStatusIdentifier': 'Result Status',
       'ResultValueTypeName': 'Result Value Type', 'LatitudeMeasure': 'Latitude', 'LongitudeMeasure':'Longitude'}
)

new_fecal_df = new_fecal_df.astype({"StateCode":int, "CountyCode":int})

#new_fecal_df.head()

state_code = new_fecal_df['State'] = new_fecal_df['StateCode'].apply(lambda x: str(x).zfill(2))
state_code

county_code = new_fecal_df['County'] = new_fecal_df['CountyCode'].apply(lambda x: str(x).zfill(3))
county_code

new_fecal_df

FIPS = new_fecal_df['FIPS'] = new_fecal_df.State + new_fecal_df.County
FIPS

new_fecal_df.columns

new_fecal_df = new_fecal_df[['FIPS', 'Org Identifier', 'Org Name', 'Start Date', 'Start Time', 'Time Zone',
       'MonitoringLocationIdentifier', 'Characteristic Name', 'Result Value',
       'Unit Code', 'Result Status', 'Result Value Type', 'Latitude',
       'Longitude', 'StateCode', 'CountyCode']]
new_fecal_df
```
