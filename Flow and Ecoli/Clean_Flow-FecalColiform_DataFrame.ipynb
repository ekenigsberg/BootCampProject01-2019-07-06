```python
import pandas as pd
```

```python
flow_csv = "Flow_Rate_Instantaneous_1972-2012.csv"
fecal_coli_csv = "Fecal_Coliform_1972-2012.csv"


flow_df = pd.read_csv(flow_csv)
fecal_df = pd.read_csv(fecal_coli_csv)

flo_df.head()
```

```python
flow_df.columns
```

```python
clean_flow_df = flow_df.dropna(axis=1, how='all')
clean_flow_df
```

```python
clean_flow_df.columns
```

```python
new_flow_df = clean_flow_df[['OrganizationIdentifier', 'OrganizationFormalName',
       'ActivityIdentifier', 'ActivityStartDate', 'ActivityStartTime/Time',
       'ActivityStartTime/TimeZoneCode', 'MonitoringLocationIdentifier',
       'ResultIdentifier','CharacteristicName', 'ResultMeasureValue',
       'ResultMeasure/MeasureUnitCode', 'ResultStatusIdentifier', 'ResultValueTypeName','USGSPCode','ProviderName']]
                            
new_flow_df
```

```python
clean_new_flow_df = new_flow_df.dropna(axis=0, how='any')
clean_new_flow_df
```

```python
clean_new_flow_df.to_csv("Clean_Flow_Data.csv", index=False)
```

```python
fecal_df.head()
```

```python
clean_fecal_df = fecal_df.dropna(axis=1, how='all')
clean_fecal_df
```

```python
clean_fecal_df.columns
```

```python
new_fecal_df = clean_fecal_df[['OrganizationIdentifier', 'OrganizationFormalName',
       'ActivityIdentifier', 'ActivityStartDate', 'ActivityStartTime/Time',
       'ActivityStartTime/TimeZoneCode', 'MonitoringLocationIdentifier',
       'ResultIdentifier','CharacteristicName','ResultMeasureValue',
       'ResultMeasure/MeasureUnitCode','ResultStatusIdentifier','ResultValueTypeName','USGSPCode','ProviderName']]

new_fecal_df
```

```python
clean_new_fecal_df = new_fecal_df.dropna(axis=0, how='any')
clean_new_fecal_df
```

```python
clean_new_fecal_df.to_csv("Clean_Fecal_Coliform_Data.csv", index=False)
```
